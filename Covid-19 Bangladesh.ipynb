```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
covid_bd=pd.read_csv('district-wise-confirmed-recovered-cases_08.06.2020.csv')
```

```python
plt.figure(figsize=(12, 7))
sns.barplot(x='Division',y='Confirmed_cases Upto 08 June',data=covid_bd)
```

```python
covid_bd.head(4)
```

```python
division=covid_bd.drop(['District_name','Dist_Code','Objectid'],axis=1)
```

```python
division.head()
```

```python
division=divison.groupby('Division').sum()
```

```python
division.head()
```

```python
df = division.transpose()
```

```python
total_cases=pd.DataFrame(df.sum())
```

```python
total_cases.head()
```

```python
total_cases.columns = ['Total Cases']
```

```python
total_cases.head(4)
```

```python
district=covid_bd.drop(['Division','Dist_Code','Objectid'],axis=1)
```

```python
district.head(3)
```

```python
district=district.groupby('District_name').sum()
```

```python
district.head(3)
```

```python
df1 = district.transpose()
```

```python
total_cases_district=pd.DataFrame(df1.sum())
```

```python
total_cases_district.head()
```

```python
total_cases_district.columns  = ['Total Cases']
```

```python
total_cases_district.head(4)
```

```python
total_cases_district
```

```python
pd.DataFrame(total_cases_district).plot(kind='bar',figsize=(15,5))
plt.ylabel('Total Cases')
plt.show()
```

```python
pd.DataFrame(total_cases).plot(kind='bar',figsize=(15,5))
plt.ylabel('Total Cases')
plt.show()
```
